## Introduction
This project implements a fun version of the classic Rock-Paper-Scissors game using Python. The game allows the user to play multiple rounds against the computer, with a scoreboard to track wins, losses, and ties. It also includes humorous messages, a dramatic countdown, and a hidden cheat code to enhance engagement.


```python
import random  # For generating the computer's random choice
import time    # For adding countdown delay effects

```

```python
def get_computer_choice(user_choice=None, cheat_mode=False):
    """
    Returns the computer's choice. If cheat_mode is enabled,
    the computer always loses to the user's choice.
    """
    if cheat_mode and user_choice:
        if user_choice == "rock":
            return "scissors"
        elif user_choice == "paper":
            return "rock"
        elif user_choice == "scissors":
            return "paper"
    return random.choice(["rock", "paper", "scissors"])


def determine_winner(user_choice, computer_choice):
    """
    Determines the winner of the round based on user and computer choices.
    """
    if user_choice == computer_choice:
        return "Tie"
    elif (
        (user_choice == "rock" and computer_choice == "scissors") or
        (user_choice == "paper" and computer_choice == "rock") or
        (user_choice == "scissors" and computer_choice == "paper")
    ):
        return "Win"
    else:
        return "Lose"
```

```python

# Initialize scoreboard
score = {"Wins": 0, "Losses": 0, "Ties": 0}
cheat_mode = False  # Flag to toggle cheat mode


# Main game loop
while True:
    # Ask for user input
    user_choice = input("Enter rock, paper, or scissors (or 'quit' to exit, 'predictor' to cheat): ").lower()
    
    if user_choice == "quit":
        print("Thanks for playing! Final Scoreboard:")
        print(score)
        break

    if user_choice == "predictor":
        cheat_mode = True
        continue

    if user_choice not in ["rock", "paper", "scissors"]:
        print("Invalid input. Try again!")
        continue

    # Countdown animation
    print("Rock...")
    time.sleep(0.5)  # Adding a delay for dramatic effect
    print("Paper...")
    time.sleep(0.5)  # Adding a delay for dramatic effect
    print("Scissors...")  
    time.sleep(0.5)  # Adding a delay for dramatic effect
    print("Shoot!")
    # Get the computer's move
    computer_choice = get_computer_choice(user_choice, cheat_mode)
    print(f"Computer chose: {computer_choice}")

    # Determine result
    result = determine_winner(user_choice, computer_choice)

    # Update scoreboard and print outcome
    if result == "Win":
        score["Wins"] += 1
        print("You crushed it! üòé")
    elif result == "Lose":
        score["Losses"] += 1
        print("Ouch, better luck next time! ü§ñ")
    else:
        score["Ties"] += 1
        print("A tie? We‚Äôre evenly matched. ü§ù")

    # Show updated scoreboard
    print(f"Scoreboard: {score}\n")
    print("Let's play again!\n")
    print("Did you know? Rock-paper-scissors is believed to have originated in China around 206 BC to 220 AD!")
```
